```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from scipy.cluster import hierarchy as hc
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv('Project 6 dataset1.csv')
```

```python
dataset["genre"]=dataset["genre"].fillna(dataset["genre"].mode()[0])
dataset["rating"]=dataset["rating"].fillna(dataset["rating"].mean())
dataset["type"]=dataset["type"].fillna(dataset["type"].mode()[0])
```

```python
dataset.dtypes
```

```python
fig,ax=plt.subplots(1,1,figsize=(20,12))
sns.heatmap(dataset.corr(),annot=True,linewidths=1)
plt.show()
```

```python
columns_to_keep = ['rating', 'genre']
X=dataset[columns_to_keep].copy()
X
```

```python
le=LabelEncoder()
X['genre']=le.fit_transform(X['genre'])
X
```

```python
columns_for_the_test=['members']
y=dataset[columns_for_the_test].copy()
y
```

```python
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
wcss=[]
for i in range(1,7):
  km=KMeans(n_clusters=i,init='k-means++',random_state=4)
  km.fit(X)
  wcss.append(km.inertia_)
```

```python
plt.plot(range(1,7),wcss)
plt.xlabel("number of cluster")
plt.ylabel("WCSS")
plt.show()
```

```python
km1=KMeans(n_clusters=3,init='k-means++',random_state=5)
y_kmeans=km1.fit_predict(X)
y_kmeans
```

```python
X_1=np.array(X)
plt.scatter(X_1[y_kmeans==0,0], X_1[y_kmeans==0,1],s = 100,c = 'violet', label = 'Cluster 1')
plt.scatter(X_1[y_kmeans==1,0], X_1[y_kmeans==1,1],s = 100,c = 'blue', label = 'Cluster 2')
plt.scatter(X_1[y_kmeans==2,0], X_1[y_kmeans==2,1],s = 100,c = 'green', label = 'Cluster 3')
plt.scatter(km1.cluster_centers_[:,0], km1.cluster_centers_[:,1], s = 100, c = 'black', label = 'Centroid')
plt.title("Clusters")
plt.legend()
plt.show()
```

```python
dendrogram=hc.dendrogram(hc.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('genre')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
ac_1=ac.fit_predict(X)
```

```python
plt.scatter(X[ac_1==0,0], X[ac_1==0,1],s = 100,c = 'violet', label = 'Cluster 1')
plt.scatter(X[ac_1==1,0], X[ac_1==1,1],s = 100,c = 'blue', label = 'Cluster 2')
plt.scatter(X[ac_1==2,0], X[ac_1==2,1],s = 100,c = 'green', label = 'Cluster 3')
plt.title("Clusters Using the hierchical Clustering")
plt.legend()
plt.show()
```
